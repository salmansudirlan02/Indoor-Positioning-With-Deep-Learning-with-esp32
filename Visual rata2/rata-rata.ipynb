```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
features = [f"feature_{i}" for i in range(128)]
df = pd.read_csv('csi_data_tw2.csv')
```

```python
X = df.drop(columns=['Titik','3','64','65'])
# X = df.drop(columns=['Titik'])
y = df['Titik']
```

```python
def standard_scaler(data):
    return (data - np.mean(data)) / np.std(data)
```

```python
for col in X.columns:
    unique_val = len(X[col].value_counts())
    # print(f"Nilai unik fitur {col}: ", unique_val)
    if unique_val < 2:
        X = X.drop(columns=col)
```

```python
for cols in X.columns:
    X[cols] = standard_scaler(X[cols])
```

```python
X
```

```python
data = pd.concat([X,y], axis=1)
data
```

```python
rata_rata_per_fitur = data.groupby('Titik').mean().reset_index()
rata_rata_per_fitur
```

```python
total_rata_rata_per_kelas = rata_rata_per_fitur.sum(axis=1)
total_rata_rata_per_kelas
```

```python
from sklearn.preprocessing import normalize
```

```python
normalized_data = normalize(X, norm='l2')
```

```python
normalized_data
normalized_X = pd.DataFrame(normalized_data, columns=X.columns)
```

```python
normalized_X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(normalized_X, y, test_size=0.2, random_state=142, stratify=y)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Flatten, Dropout, Bidirectional

num_classes = 10

# Membuat model
model = Sequential()

# Layer CNN1D
model.add(Conv1D(filters=128, kernel_size=3, strides=2, activation='relu', input_shape=(125,1)))
model.add(Conv1D(filters=128, kernel_size=3, strides=1, activation='relu'))
model.add(MaxPooling1D(pool_size=2))

# Layer LSTM
model.add(Bidirectional(LSTM(128, return_sequences=True)))
model.add(Bidirectional(LSTM(128)))

# Layer DNN
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=100, batch_size=128, validation_data=(X_val, y_val))
```
