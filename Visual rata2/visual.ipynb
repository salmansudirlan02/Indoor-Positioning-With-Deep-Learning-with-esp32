```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from CSIKit.filters.passband import lowpass
from CSIKit.filters.statistical import running_mean
from CSIKit.util.filters import hampel

# Load the CSV file
file_path = 'cleaned_file_titik0.csv'
csi_data_df = pd.read_csv(file_path)

# Assuming 'CSI_DATA' contains string representations of arrays
csi_data_str = csi_data_df['CSI_DATA'].apply(lambda x: np.fromstring(x[1:-1], sep=' '))

# Convert to complex numbers and then calculate amplitudes for each pair
csi_amplitudes = csi_data_str.apply(lambda arr: np.abs(np.array([complex(arr[i], arr[i+1]) for i in range(0, len(arr), 2)])))

# Filter and process the data
filtered_csi = []
for frame in csi_amplitudes:
    if isinstance(frame, np.ndarray) and frame.ndim == 1:
        frame_filtered = lowpass(frame, 10, 100, 5)
        frame_filtered = hampel(frame_filtered, 10, 3)
        frame_filtered = running_mean(frame_filtered, 10)
        filtered_csi.append(frame_filtered)

# Convert the list of filtered frames into a numpy array for visualization
csi_matrix_filtered = np.array(filtered_csi)

# Check if csi_matrix_filtered is empty
if csi_matrix_filtered.size == 0:
    print("No data available for visualization.")
else:
    # Visualize the filtered data
    plt.figure(figsize=(15, 5))
    plt.imshow(csi_matrix_filtered.T, aspect='auto', cmap='viridis', interpolation='none')
    plt.colorbar(label='Amplitude', extend='max')
    plt.clim(0, 100)
    plt.title("CSI Amplitude Heatmap After Filtering")
    plt.xlabel("Frame")
    plt.ylabel("Subcarrier")
    plt.show()
```
