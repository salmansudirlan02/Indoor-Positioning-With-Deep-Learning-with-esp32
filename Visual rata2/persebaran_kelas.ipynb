```python
import pandas as pd

# Membaca file CSV yang diunggah
file_path = 'csi_data_tw2.csv'
data = pd.read_csv(file_path)

# Menampilkan beberapa baris pertama untuk memahami struktur data
data.head()
```

# KELAS 0 PADA DATASET TOWER 2 TIDAK DIPAKAI

```python
# KELAS 0 PADA DATASET TOWER 2 TIDAK DIPAKAI

import matplotlib.pyplot as plt

# Menghitung frekuensi untuk setiap kelas dalam kolom 'Titik'
class_counts = data['Titik'].value_counts()

# Membuat bar plot
plt.figure(figsize=(12, 6))
class_counts.plot(kind='bar')
plt.title('Distribusi Kelas dalam Data')
plt.xlabel('Kelas')
plt.ylabel('Jumlah Pengamatan')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import seaborn as sns

# Membuat subplots
fig, axes = plt.subplots(1, 3, figsize=(20, 5))

# Membuat boxplot pada setiap subplot
sns.boxplot(data=data[["1", "2", "3"]], ax=axes[0])
sns.boxplot(data=data[["4", "5", "6"]], ax=axes[1])
sns.boxplot(data=data[["7", "8", "9"]], ax=axes[2])  # Memastikan kolomnya "7", "8", "9"
```
