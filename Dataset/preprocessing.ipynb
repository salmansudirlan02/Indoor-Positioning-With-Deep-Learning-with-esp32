```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
```

```python
# Inisialisasi DataFrame kosong untuk menyimpan semua data
all_data = pd.DataFrame()

# Loop melalui setiap file
for i in range(0, 10):  # Ini akan loop dari 1 sampai 9
    file_name = f'titik{i}_ruijie.csv'
    # Baca file
    temp_data = pd.read_csv(file_name)
    temp_csi = temp_data['CSI_DATA']
    cleaned_file_path = f'cleaned_file_titik{i}.csv'
    temp_csi.to_csv(cleaned_file_path, index=False)

# Sekarang, all_data berisi data dari semua file
```

```python
def parse_csi_string(csi_string):
    return [int(x) for x in csi_string.strip('[]').split()]
```

```python
all_data = pd.DataFrame()

for i in range(0, 10):  # Ini akan loop dari 0 sampai 9
    file_path = f'cleaned_file_titik{i}.csv'
    # Baca file
    data = pd.read_csv(file_path, converters={0: parse_csi_string})
    parsed_data = pd.DataFrame(data.iloc[:, 0].tolist())
    parsed_data['Titik'] = i
    all_data = pd.concat([all_data, parsed_data], ignore_index=True)
```
